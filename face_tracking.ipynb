```python
%matplotlib widget
import matplotlib.pyplot as plt
import cv2
import pyautogui as robot
```

```python
cam=cv2.VideoCapture(0)
```

```python

face_model = cv2.CascadeClassifier(r'D:\python_code\rezomeh\poshe 4\haarcascade_frontalface_default.xml')
face_model1 = cv2.CascadeClassifier(r'D:\python_code\rezomeh\poshe 4\haarcascade_eye.xml')
```

```python
face_model
```

```python
if face_model.empty():
    print("Error loading face cascade classifier!")
```

```python
loop = True
cam=cv2.VideoCapture(0)
while loop:
    # خواندن تصویر از دوربین
    _, img = cam.read()
    img=cv2.flip(img,1)
    
    # تبدیل به سطح خاکستری
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # پیش‌پردازش تصویر
    gray = cv2.equalizeHist(gray)
    gray = cv2.GaussianBlur(gray, (5, 5), 0)
    
    # شناسایی چهره
    face = face_model.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    if len(face) > 0:  # بررسی وجود حداقل یک چهره
        x, y, w, h = face[0]
        x2, y2 = x + w, y + h
        
        # استخراج بخش چهره
        gray_face = gray[y:y2, x:x2]
        
        # شناسایی چشم‌ها در چهره
        eye = face_model1.detectMultiScale(gray_face, scaleFactor=1.1, minNeighbors=5, minSize=(15, 15))
        
        # رسم مستطیل دور چهره
        imgout = img.copy()
        out = cv2.rectangle(imgout, (x, y), (x2, y2), (0, 255, 0), 2)

        white=(250,250,250)
        red=(0,0,255)
        range=white
        mouse_x=robot.position().x
        mouse_y=robot.position().y
        
        if y<80:
            range=red
            mouse_y=mouse_y-abs(y-80)
            robot.moveTo(mouse_x,mouse_y,0.4)
        
        if y2>350:
            range=red
            mouse_y=mouse_y+abs(y-200)
            robot.moveTo(mouse_x,mouse_y,0.4)
        
        
        if x<200:
            range=red
            mouse_x=mouse_x-abs(x-200)
            robot.moveTo(mouse_x,mouse_y,0)
        
        
        
        if x2>450:
            range=red
            mouse_x=mouse_x+abs(x-450)
            robot.moveTo(mouse_x,mouse_y,0)

        
        out = cv2.rectangle(imgout, (200,80), (450,350), range, 2)

        
        ic = 0  # شمارش چشم‌ها
        for (xe, ye, ew, eh) in eye:
            ic += 1
            cv2.rectangle(out, (x + xe, y + ye), (x + xe + ew, y + ye + eh), (0, 0, 255), 2)
            if ic == 2:  # اگر دو چشم شناسایی شدند، متوقف شود
                break
        
        # نمایش تصویر
        cv2.imshow('image', out)
    
    else:
        print("No faces detected.")
    
    # شرط خروج از حلقه
    if cv2.waitKey(1) == ord("q"):
        loop = False

# بستن دوربین و پنجره‌ها
cam.release()
cv2.destroyAllWindows()
```
